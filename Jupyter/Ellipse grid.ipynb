```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

t = np.linspace(-np.pi,np.pi,100)   #parameter for ellipse  
sma = 0.5     #Semi major axis
smi = 0.15    #Semi minor axis


x = sma*np.cos(t)
y = smi*np.sin(t)


inter=50
a = np.linspace(-sma,sma,inter)
b = np.linspace(-sma,sma,inter)

aa,bb = np.meshgrid(a,b)
ell = (aa**2/sma**2) + (bb**2/smi**2)




rows,col = ell.shape
ell_x = np.array([])
ell_y = np.array([])



for i in range(rows):
    for j in range(col):
        if(ell[i,j]<=1):
            ell_x = np.append(ell_x,aa[i,j])
            ell_y = np.append(ell_y,bb[i,j])
 


size = ell_x.size
sample = np.linspace(0,100,size)
# Maximum Possible rectangle that fits inside an ellipse
# l = 2*sma/(2**(1/2))
# b = 2*smi/(2**(1/2))

# l = ((sma**2 + smi**2)**(1/2))/sma*smi
# b = ((sma**2 + smi**2)**(1/2))/sma*smi

# rec_x = [-l/2,l/2,l/2,-l/2]
# rec_y = [-b/2,-b/2,b/2,b/2]



fig,ax = plt.subplots()
ax = plt.axes(projection='3d')
ax.scatter(ell_x,ell_y,sample)
# ax.plot(x,y)
# ax.scatter(ell_x,ell_y)
# ax.scatter(rec_x,rec_y)
```
